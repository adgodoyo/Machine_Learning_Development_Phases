```python
import pandas as pd
from PIL import Image
from ultralytics import YOLO

# Cargamos un modelo preentrenado YOLOv8n
model = YOLO('yolov8n.pt')

# Carga tu imagen
results = model('Ejemplos 1/20220308_190122.jpg')  # results list

# Inicializar una lista para almacenar los resultados
detections = []

# Iterar sobre los resultados de la detección
for r in results:
    # Extraer las etiquetas detectadas
    for box in r.boxes:
        class_id = int(box.cls)  # ID de la clase detectada
        label = model.names[class_id]  # Etiqueta correspondiente a la clase
        confidence = box.conf.item()  # Convertir la confianza a un escalar
        
        # Extraer las coordenadas del cuadro delimitador (bounding box)
        xywh = box.xywh[0].cpu().numpy()  # Coordenadas en formato (x, y, w, h)
        
        # Agregar los resultados a la lista
        detections.append({
            "Etiqueta": label,
            "Confianza": confidence,
            "Centro X": xywh[0],
            "Centro Y": xywh[1],
            "Ancho": xywh[2],
            "Alto": xywh[3]
        })

# Convertir la lista a un DataFrame de pandas
df_detections = pd.DataFrame(detections)

# Mostrar el DataFrame
print(df_detections)
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Open the video file
video_path = "Ejemplos 2/20221225_143835.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import pandas as pd
from ultralytics import YOLO

# Cargar el modelo YOLOv8 preentrenado
model = YOLO('yolov8n.pt')

# Ruta del video
video_path = "Ejemplos 2/20221225_143835.mp4"
cap = cv2.VideoCapture(video_path)

# Lista para almacenar todas las detecciones
all_detections = []

# Contador de frames
frame_id = 0

# Procesar cada frame del video
while cap.isOpened():
    # Leer un frame del video
    success, frame = cap.read()

    if success:
        frame_id += 1  # Incrementar el contador de frames
        
        # Ejecutar la inferencia de YOLOv8 en el frame
        results = model(frame)

        # Extraer las detecciones del frame
        for r in results:
            for box in r.boxes:
                class_id = int(box.cls)  # ID de la clase detectada
                label = model.names[class_id]  # Etiqueta de la clase
                confidence = box.conf.item()  # Convertir la confianza a escalar
                
                # Extraer las coordenadas del cuadro delimitador
                xywh = box.xywh[0].cpu().numpy()  # Coordenadas en formato (x, y, w, h)
                
                # Almacenar la información en la lista
                all_detections.append({
                    "Frame": frame_id,
                    "Etiqueta": label,
                    "Confianza": confidence,
                    "Centro X": xywh[0],
                    "Centro Y": xywh[1],
                    "Ancho": xywh[2],
                    "Alto": xywh[3]
                })
        
        # Visualizar los resultados en el frame
        annotated_frame = results[0].plot()
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Romper el loop si se presiona 'q'
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Romper el loop si el video termina
        break

# Convertir las detecciones a un DataFrame
df_detections = pd.DataFrame(all_detections)

# Mostrar el DataFrame
print(df_detections)

# Guardar el DataFrame en un archivo CSV (opcional)
df_detections.to_csv('detecciones_video.csv', index=False)

# Liberar el objeto de captura de video y cerrar las ventanas
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import time

# Carga el modelo YOLOv8
model = YOLO('yolov8n.pt')

# Inicializa la captura de video para la cámara en vivo.
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Camera is not enabled.")
    exit()

# Tiempo límite para la autenticación
time_limit = time.time() + 15  # 15 segundos desde ahora

# Bucle para procesar los cuadros capturados por la cámara en tiempo real
def process_frames():
    # Limpia el archivo para la próxima iteración
    with open('resultados.txt', 'w') as archivo:
        pass
    while time.time() < time_limit:
        success, frame = cap.read()
        
        if success:
            results = model(frame)
            
            # Guarda los resultados de la detección en un archivo txt
            for r in results:
                r.save_txt("resultados.txt", save_conf=False)

            annotated_frame = results[0].plot()
            cv2.imshow("Inferencia YOLOv8", annotated_frame)
            cv2.waitKey(1)  # Necesario para que la ventana de OpenCV funcione correctamente
            
            with open('resultados.txt', 'r') as archivo:
                persona_detectada = False
                for linea in archivo:
                    datos = linea.split()
                    if datos and datos[0] == "0":  # Suponiendo que '0' es la etiqueta para personas
                        persona_detectada = True
                        break

                if persona_detectada:
                    print("Persona autenticada")
                    return
                else:
                    print("No persona detectada. Intente de nuevo.")

            
        
        else:
            print("La imagen no se recibió, intente de nuevo")
    
    print("Autenticación fallida")

process_frames()

# Libera el objeto de captura de video y cierra todas las ventanas de OpenCV
cap.release()
cv2.destroyAllWindows()

```
